```python
import json

with open('test.json', 'r') as f:
    data = json.load(f)

print(data[0:5])
```

```python
print(type(data))
```

```python
for name in data:
    print(name)
```

```python
liste_json = []
for rum in data:
    for k, v in rum.items():
        v = v.replace(' cl', 'cl')
        liste_json.append(v)
liste_json
```

```python
values = str(liste_json[0]).split()
values
```
